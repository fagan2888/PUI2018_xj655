```python
bk_tmp = bk
bk_tmp = bk_tmp.dropna()
#have to drop nan, otherwise I will not able to transfer the format(Nan is infinit, a special part)
bk_tmp['LandUse'] = bk_tmp['LandUse'].astype(int)
bkgroup = bk_tmp.groupby('LandUse').count()
bkgroup.head(11)
```

```python
mn.groupby('LandUse').count()
```

```python
mn_tmp = mn
mn_tmp = mn_tmp.dropna()
#have to drop nan, otherwise I will not able to transfer the format(Nan is infinit, a special part)
mn_tmp['LandUse'] = mn_tmp['LandUse'].astype(int)
mngroup = mn_tmp.groupby('LandUse').count()
mngroup.head(10)
```

```python
# merge the dataframes on LandUse. You can use the pandas function merge ( _pd.merge)
both = pd.merge(bkgroup,mngroup,left_index=True,right_index=True).reset_index()
both.head()
#reset index will not cover landuse
```

```python
both.columns
```

```python
both1 = both.rename(index=str, columns={"LandUse": "LandUse", "ZipCode_x": "bk","ZipCode_y":"mn"})
both1
```

```python
#both_rename = pd.DataFrame({"LandUse": ['One &Two Family Buildings', 'Multi-Family Walk-Up Buildings', 'Multi-Family Elevator Buildings', 'Mixed Residential & Commercial Buildings', 'Commercial & Office Buildings', 'Industrial & Manufacturing', 'Transportation & Utility', 'Public Facilities & Institutions','Open Space & Outdoor Recreation', 'Parking Facilities', 'Vacant Land']})
#both1.update(both_rename)
'''I don't know why it doesn't work 
'''

both1['LandUse']= ['One &Two Family Buildings', 'Multi-Family Walk-Up Buildings', 'Multi-Family Elevator Buildings', 'Mixed Residential & Commercial Buildings', 'Commercial & Office Buildings', 'Industrial & Manufacturing', 'Transportation & Utility', 'Public Facilities & Institutions','Open Space & Outdoor Recreation', 'Parking Facilities', 'Vacant Land',]
both1.LandUse
both1
```

```python
# calculate the ratio of BK to Man buildings per LandUse category

both1['luratio'] = both1.bk/both1.mn
both1
```

Summary: The building in Manhattan landuse for Multi-Family Walk-Up Buildings is more than Brooklyn.

```python
def Manhwin(luratio):
    if luratio < 1:
        return True
    else:
        return False
    
both1['Manhattan_Wins'] = both1.luratio.apply(Manhwin)
#new column has to add 
both1
```

Only the land use of buildings in Manhattan  for Multi-Family Walk-Up Buildings, has more counts than Brooklyn. Maybe because the house in Manhattan is old and has no elevator and a lot of people live there. Many family share one building.
